```python
import pandas as pd
```

```python
diamonds_df = pd.read_csv(path_to_file)
```

```python
diamonds_df
```

```python
X = diamonds_df.drop(['price'],axis=1)
y = diamonds_df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred_train = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# R2 adjusted
def adjusted_r2_score(r2, n, k):
    adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - k - 1))
    return adjusted_r2

import numpy as np
def print_metrics(df,y_true,y_pred):
  mse = mean_squared_error(y_true, y_pred)
  r2 = r2_score(y_true, y_pred)
  mae = mean_absolute_error(y_true, y_pred)
  print(f"MSE: {mse}")
  print(f"R2: {r2}")
  print(f"MAE: {mae}")
  print(f"RMSE: {np.sqrt(mse)} ")
  adj_r2 = adjusted_r2_score(r2,len(df),len(df.columns))
  print(f"adj_r2: {adj_r2}")

print_metrics(diamonds_df,y_test,y_pred_train)
```
