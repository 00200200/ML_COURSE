```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/CL/Day 4/imdb_reviews_preprocessed_2.fixed.csv')
df
```

```python
df['label'] = df['label'].apply(lambda sentiment: 1 if sentiment == 'pos' else 0)
```

```python
df
```

```python
df['review_lemmatized'] = df['review_lemmatized'].apply(lambda x : x.split())
```

```python
df
```

```python
from gensim.models import Word2Vec

model = Word2Vec(df['review_lemmatized'],vector_size=100,window=2,min_count=3,workers=3,sg=0,epochs=10)
```

```python
words = ['cast', 'movie', 'comedy', 'watch', 'interesting']
for word in words:
  if word in model.wv:
    print(f"Mo st similar dla slowa {word} : ${model.wv.most_similar(word,topn=10)} '\n'")
```

```python
#4
print("\n")
print(f"podobienstwo movie do comedy ")
print(model.wv.similarity("movie", "comedy"))
```

```python
#5
print("\n wektor dla show")
if 'show' in model.wv:
  print(model.wv['show'])
print("\n")
```

```python
#6
df['mean_vector'] = df['review_lemmatized'].apply(lambda x: model.wv.get_mean_vector(x))
df
```

```python
df
```

```python
from sklearn.linear_model import LogisticRegression
lreg = LogisticRegression(max_iter=1234)
from sklearn.model_selection import train_test_split

X = np.array(list(df['mean_vector'].values))
y = df['label']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

lreg.fit(X_train,y_train)
y_pred = lreg.predict(X_test)

from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred))
```
