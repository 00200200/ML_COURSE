```python
path_to_file = 'drive/MyDrive/CL/Day_1/kc_house_data_preprocessed.csv'
import pandas as pd
diamonds_df = pd.read_csv(path_to_file)
```

```python
diamonds_df
```

```python
diamonds_df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
X = diamonds_df.drop(['price'],axis=1)
y = diamonds_df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred_train = model.predict(X_test)

```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

import numpy as np
def print_metrics(df,y_true,y_pred):
  mse = mean_squared_error(y_true, y_pred)
  r2 = r2_score(y_true, y_pred)
  mae = mean_absolute_error(y_true, y_pred)
  print(mse)
  print(r2)
  print(mae)
  print(np.sqrt(mse))

print_metrics(diamonds_df,y_test,y_pred_train)
```
