```python
obj_columns = df.select_dtypes(include=['object']).columns
obj_columns
```

```python
for col in obj_columns:
  df[col] = df[col].astype('category')
df.info(
)
```

```python
cat_columns = df.select_dtypes(include=['category']).columns
cat_columns
```

```python
df[cat_columns] = df.str.replace
```

```python
#FBS,EXANG,AHD 0 - 1 GetDummies Drop First for the rest GetDummiesWithoutDropFirst
cat_df = ['fbs','exang','ahd']
cat_wdf = ['sex', 'chestpain',  'thal']
df
df = pd.get_dummies(data=df,columns=cat_df,drop_first=True,dtype=int) # NIE TAK
```

```python
df['fbs_Yes'] = df['fbs_Yes'].astype(int)
df['exang_Yes'] = df['exang_Yes'].astype(int)
df['ahd_Yes'] = df['ahd_Yes'].astype(int)
```
