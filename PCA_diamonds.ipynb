```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/CL/Day_5_6/diamonds_preprocessed.csv')
df
```

```python
X = df.drop(['price'],axis=1)
y = df['price']
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X = df.drop(['price'],axis=1)
y = df['price']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

lreg = LinearRegression()
lreg.fit(X_train,y_train)
y_pred = lreg.predict(X_test)
from sklearn.metrics import r2_score

print(r2_score(y_test,y_pred))
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
for i in range(len(df.columns) -1):
  pca = PCA(n_components=i,random_state=42)
  pca.fit_transform(X)
  print(pca.explained_variance_ratio_)
```

```python
pca = PCA(n_components=0.80,random_state=42)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_reduced = pca.fit_transform(X_scaled)
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
diamonds_pca_df = pd.DataFrame(X_reduced,columns=['PC1','PC2','PC3','PC4'])
diamonds_pca_df['price'] = y
```

```python
diamonds_pca_df
```

```python
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split

X = diamonds_pca_df.drop(['price'],axis=1)
y = diamonds_pca_df['price']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

lreg = LinearRegression()
lreg.fit(X_train,y_train)
y_pred = lreg.predict(X_test)
from sklearn.metrics import r2_score

print(r2_score(y_test,y_pred))
```
