```python
corr_variables = ['sqft_living','grade','sqft_above','sqft_living15']

for var in corr_variables:
  plt.figure(figsize=(8,6))
  sns.scatterplot(x=housing_df[var],y=housing_df['price'])
  plt.title(var)
  plt.xlabel(var)
  plt.ylabel('price')
  plt.show()
```

```python

# IMPLEMETNIG IQR METHOD
# [Q1 - 1.5 * IQR , x , Q3+1.5*IQR]
#IQR - ROZSTEP CWIQRTKOWY CZYLI ROZNICA MIEDYZ TRZECIM A PIERWSZYM KWARTYLEM
def remove_outliners(df,column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3-Q1
  lv = Q1 - 1.5 * IQR
  uv = Q3 + 1.5 * IQR
  return df[(df[column] >= lv) & (df[column] <= uv)]
print(housing_df.shape)
for column in corr_variables:
  housing_df = remove_outliners(housing_df,column)
  print(housing_df.shape)

```

```python
from sklearn.model_selection import train_test_split
x = housing_df.drop('price',axis=1)
y = housing_df['price']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print(f"R2 score: {lr.score(x_test,y_test)}")
print(f"Alternative method: {r2_score(y_test,y_pred)}")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model_scaled = LinearRegression()
model_scaled.fit(x_train_scaled,y_train)
y_pred = model_scaled.predict(x_test_scaled)
print(r2_score(y_test,y_pred))
```
