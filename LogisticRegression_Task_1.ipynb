```python
import pandas as pd

lumbar_df = pd.read_csv('/content/drive/MyDrive/CL/Day_2/lumbar_df_normalized.csv')
```

```python
lumbar_df.head(10)
lumbar_df.duplicated().sum()
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
X = lumbar_df.drop(['class'],axis=1)
y = lumbar_df['class']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
y_pred = logreg.predict(X_test)
from sklearn.metrics import precision_score
precision = precision_score(y_test,y_pred)
precision
```

```python
import numpy as np
params = [
    {'penalty':['l1','l2','elasticnet','none'],
    'C' : np.logspace(-4,4,20),
    'solver': ['lbfgs','newton-cg','liblinear','sag','saga'],
    'max_iter'  : [100,1000,2500,5000]
}
]
gscv = GridSearchCV(LogisticRegression(),param_grid=params,cv=4,scoring='precision')
gscv.fit(X_train,y_train)
gscv.best_params_
```

```python
model = gscv.best_estimator_
print(gscv.best_params_)
```

```python
y_pred = model.predict(X_test)
lreg = LogisticRegression(C=1.3, class_weight='balanced', max_iter=100, penalty='l1', solver='liblinear')
lreg.fit(X_train,y_train)
y_pred = lreg.predict(X_test)
print(precision_score(y_test,y_pred))
print(precision_score(y_train,lreg.predict(X_train)))
```
