```python
# Wczytaj zbiór danych o diamentach, który był wynikiem zadania "Regresja liniowa — zadanie 2" — diamonds_preprocessed.csv.
diamonds_df = pd.read_csv('drive/MyDrive/CL/Day_1/diamonds_preprocessed.csv')
diamonds_df
```

```python
# Podziel zbiór danych na treningowy i testowy.
X = diamonds_df.drop(['price'],axis=1)
y = diamonds_df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Wykorzystując odpowiednie narzędzia biblioteki Scikit-learn, wygeneruj zmienne wielomianowe. Dobierz odpowiedni stopień wielomianu.
# Dla każdego wariantu:
# wytrenuj model regresji liniowej korzystając z nowego zestawu zmiennych
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.linear_model import Ridge,Lasso
import numpy as np
def print_metrics(y_true,y_pred,degree):
  r2 = r2_score(y_true,y_pred)
  mse = mean_squared_error(y_true,y_pred)
  mae = mean_absolute_error(y_true,y_pred)
  rmse = np.sqrt(mse)
  print(f"Degree: {degree} R2: {r2} MSE: {mse} MAE: {mae} RMSE: {rmse}")

degrees = [1,2,3,4,5]
alphas = [0,0.001,0.01,0.1,1,10,100,1000]
for degree_v in degrees:
  poly = PolynomialFeatures(degree=degree_v)
  X_poly = poly.fit_transform(X)
  x_train,x_test,y_train,y_test = train_test_split(X_poly,y,test_size=0.3)
  lreg = LinearRegression()
  lreg.fit(x_train,y_train)
  y_pred = lreg.predict(x_test)
  print_metrics(y_test,y_pred,degree_v)
```

```python
# Wykorzystując odpowiednie narzędzia biblioteki Scikit-learn, wygeneruj zmienne wielomianowe. Dobierz odpowiedni stopień wielomianu.
# Dla każdego wariantu:
# wytrenuj model regresji liniowej korzystając z nowego zestawu zmiennych
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.linear_model import Ridge,Lasso
import numpy as np
def print_metrics(y_true,y_pred,degree):
  r2 = r2_score(y_true,y_pred)
  mse = mean_squared_error(y_true,y_pred)
  mae = mean_absolute_error(y_true,y_pred)
  rmse = np.sqrt(mse)
  print(f"Degree: {degree} R2: {r2} MSE: {mse} MAE: {mae} RMSE: {rmse}")

degrees = [1,2,3,4]
alphas = [0,0.001,0.01,0.1,1,10,100,1000]
for degree_v in degrees:
  for alpha_v in alphas:
    print("-"*30)
    print(f"====Ridge(alpha={alpha_v}) ====")
    ridge = Ridge(alpha=alpha_v)
    poly = PolynomialFeatures(degree=degree_v)
    X_poly = poly.fit_transform(X)
    x_train,x_test,y_train,y_test = train_test_split(X_poly,y,test_size=0.3)
    ridge.fit(x_train,y_train)
    y_pred = ridge.predict(x_test)
    print_metrics(y_test,y_pred,degree_v)
```
