```python
import pandas as pd
lumbar_df = pd.read_csv('/content/drive/MyDrive/CL/Day_2/lumbar_df_normalized.csv')
lumbar_df.head(10)
```

```python
X = lumbar_df.drop(['class'],axis=1)
y = lumbar_df['class']
from sklearn.model_selection import train_test_split,GridSearchCV
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
from sklearn.tree import DecisionTreeClassifier
```

```python
params = [{
    'criterion': ['gini','entropy','log_loss'],
    'splitter': ['best','random'],
    'max_depth': [i for i in range(2,20)],
    'min_samples_split': [2, 5, 10,15,20,25,30],
    'min_samples_leaf': [i for i in range(2,20)]
}]
gscv = GridSearchCV(DecisionTreeClassifier(),param_grid=params,cv=4,scoring='precision')
gscv.fit(X_train,y_train)
```

```python
gscv.best_params_
```

```python
model = gscv.best_estimator_
y_pred = model.predict(X_test)
from sklearn.metrics import precision_score

precision = precision_score(y_test,y_pred)
print(precision)
```
