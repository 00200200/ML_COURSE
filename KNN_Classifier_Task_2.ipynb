```python
from sklearn.preprocessing import LabelEncoder
labelEncoder = LabelEncoder
columns_to_encode = [ 'island', 'sex']

# Inicjalizujemy LabelEncoder
labelEncoder = LabelEncoder()

# Funkcja pomocnicza do kodowania kolumn
def encode_column(column):
    return labelEncoder.fit_transform(column)

# Zastosowanie funkcji do wybranych kolumn
penguins_df[columns_to_encode] = penguins_df[columns_to_encode].apply(encode_column)
```

```python
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
penguins_df.dropna(inplace=True)
X = penguins_df.drop(['species'],axis=1)
y = penguins_df['species']

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
from sklearn.model_selection import GridSearchCV
params = [{
    'n_neighbors' :[ i for i in range(3,30,2)],
    'weights' : ['uniform','distance'],
    'leaf_size' : [10,20,30,40,50,60]}]

gscv = GridSearchCV(KNeighborsClassifier(),param_grid=params,cv=4,scoring='accuracy')
gscv.fit(x_train,y_train)
gscv.best_params_
```

```python
model = gscv.best_estimator_
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,model.predict(x_test))
```
