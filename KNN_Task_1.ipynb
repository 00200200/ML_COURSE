```python
lumbar_df['class'] = lumbar_df['class'].map({'Normal':0,'Hernia':1,'Spondylolisthesis':1})
lumbar_df
```

```python
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
x = lumbar_df.drop(['class'],axis=1)
y = lumbar_df['class']
x_normalized = normalizer.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_normalized,y,test_size=0.25,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
print(f"KNN SCORE {knn.score(x_test,y_test)}")
```
