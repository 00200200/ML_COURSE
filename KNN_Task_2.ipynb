```python
import pandas as pd
path_to_file = '/drive/MyDrive/CL/Day_2/penguins_size.csv'
penguins_df = pd.read_csv(path_to_file)
penguins_df
```

```python
penguins_df.dropna(inplace=True)
```

```python
#2
penguins_df.info()
obj_columns = penguins_df.select_dtypes(include=['object'])
obj_columns
for col in obj_columns:
  penguins_df[col] = penguins_df[col].astype('category')
```

```python
penguins_df.info()
```

```python
#4
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
penguins_df.drop(['island','sex'],axis=1,inplace=True)
x = penguins_df.drop(['species'],axis=1)
y = penguins_df['species']
x_normalized = normalizer.fit_transform(x)
```

```python
#5
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(x_normalized,y,test_size=0.3,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
print(f"KNN SCORE : {knn.score(x_test,y_test)}")
```
