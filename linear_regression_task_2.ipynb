```python
path_to_file = '/drive/MyDrive/diamonds.csv'
import pandas as pd

diamonds_df = pd.read_csv(path_to_file)
diamonds_df
```

```python
diamonds_df = diamonds_df.drop(columns=['Unnamed: 0'])
```

```python
print(diamonds_df[diamonds_df.duplicated()].columns)
diamonds_df
```

```python
diamonds_df.info()
diamonds_df.value_counts()
diamonds_df = pd.get_dummies(diamonds_df,columns=['cut','color','clarity'],drop_first=True)
diamonds_df
```

```python
corr = diamonds_df.corr()
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))
sns.heatmap(corr,annot=True,fmt='.2f',cmap='coolwarm')
```

```python
selected_features = corr[corr["price"].abs() > 0.3].index
selected_features
```

```python
for column in selected_features.tolist():
  plt.figure(figsize=(8,6))
  sns.histplot(diamonds_df[column])
  plt.show()
  sns.scatterplot(x=diamonds_df[column],y=diamonds_df["price"])
```

```python
def remove_outliners(df,column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3-Q1
  lv = Q1 - 1.5 * IQR
  uv = Q3 + 1.5 * IQR
  return df[(df[column] >= lv) & (df[column] <= uv)]
selected_features = selected_features.tolist()
selected_features.remove('price')
for feature in selected_features:
  remove_outliners(diamonds_df,feature)
```

```python
for column in selected_features:
  plt.figure(figsize=(8,6))
  sns.histplot(diamonds_df[column])
  plt.show()
  sns.scatterplot(x=diamonds_df[column],y=diamonds_df["price"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
x = diamonds_df[selected_features]
y = diamonds_df['price']
lr = LinearRegression()
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
print(f"score: {r2_score(y_test,y_pred)}")
```

```python

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)

model_scaled = LinearRegression()
model_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = model_scaled.predict(X_test_scaled)

r2_scaled = r2_score(y_test, y_pred_scaled)
print("Score :", r2_scaled)
```
