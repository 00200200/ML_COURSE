```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/CL/Day_5_6/blobs_dataset.csv')
```

```python
df
```

```python
from sklearn.neighbors import LocalOutlierFactor
lof = LocalOutlierFactor(n_neighbors=5,contamination=0.2)
```

```python
y_pred = lof.fit_predict(df)
y_pred
```

```python
import numpy as np
import matplotlib.pyplot as plt
lof_scores = lof.negative_outlier_factor_
marker_colors = np.array(["r", "b"])
lof_scores
```

```python
X = np.array(df)
plt.scatter(X[:, 0], X[:, 1], s=25, color=marker_colors[(y_pred + 1) // 2])
plt.title('Outliers (red) in dataset')
plt.show()
```

```python
# Teraz 5. % obserwacji
lof = LocalOutlierFactor(n_neighbors=4,contamination=0.05)
y_pred = lof.fit_predict(df)
X = np.array(df)
plt.scatter(X[:, 0], X[:, 1], s=25, color=marker_colors[(y_pred + 1) // 2])
plt.title('Outliers (red) in dataset')
plt.show()
```
