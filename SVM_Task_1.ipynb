```python
import pandas as pd
```

```python
lumbar_df = pd.read_csv('/content/drive/MyDrive/CL/Day_2/lumbar_df_normalized.csv')
lumbar_df.head(10)
```

```python
X = lumbar_df.drop(['class'],axis=1)
y = lumbar_df['class']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
from sklearn.svm import SVC
svc = SVC()

svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)
from sklearn.metrics import precision_score
print(precision_score(y_test,y_pred))
params = [{'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf'],
          'decision_function_shape' : ['ovo','ovr']
        }]
gscv = GridSearchCV(SVC(),param_grid=params,scoring='precision')
```

```python
gscv.fit(X_train,y_train)
```

```python
gscv.best_params_
```

```python
model = gscv.best_estimator_
```

```python
print(precision_score(y_test,model.predict(X_test)))
```
