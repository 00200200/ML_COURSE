```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_moons
data,labels = make_moons(500,noise=0.1,random_state=42)
data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
data_scaled
```

```python
import matplotlib.pyplot as plt
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], color='green', marker='o', s=30)
plt.title("Our dataset")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.2,min_samples=25)
dbscan.fit(data)
```

```python

labels = dbscan.labels_

plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=labels, marker='o', s=25)
plt.title("Dataset after DBSCAN clustering")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
dbscan = DBSCAN(eps=0.25, min_samples=10)
```

```python
dbscan.fit(data)

labels = dbscan.labels_
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=labels, marker='o', s=25)
plt.title("Dataset after DBSCAN clustering")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
