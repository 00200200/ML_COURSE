```python
def calculate(a:int, b:int,math_operation:str) -> int:
  if math_operation == 'add':
    return a+b
  elif math_operation == 'sub':
    return a-b
  elif math_operation == 'mul':
    return a*b
  elif math_operation =='div':
    return a/b
  else:
     raise ValueError("No math Operations")
print(calculate(3,5,'ads'))
```

```python
from typing import List
def retName(names:List[str]) -> List[str]:
  newNames: List[str] = []
  for name in names:
    if name.startswith('C'):
      newNames.append(name)
  return newNames
royal_family_names = ["Elizabeth", "Phillip", "Charles", "Camilla", "William", "Catherine", "Anne", "Harry", "Meghan"]
print(retName(royal_family_names))
```

```python
from typing import List
def retName(names:List[str]) -> List[str]:
  return list(filter(lambda name: name.startswith('C'),names))


royal_family_names = ["Elizabeth", "Phillip", "Charles", "Camilla", "William", "Catherine", "Anne", "Harry", "Meghan"]
print(retName(royal_family_names))
```

```python
from typing import Dict,Tuple

def retEmployees(employeesData:Dict[str,Tuple[str,int]],salary:int) -> Dict[str,Tuple[str,int]]:
  keys_to_remove = [key for key,value in employeesData.items() if value[1] <= salary]
  for key in keys_to_remove:
      del employeesData[key]
  return employeesData

employeeData = {
    'name1 surname1' : ('1',3000),
    'name2 surname2' : ('1',4000),
    'name3 surname3' : ('1',5000)
}

print(retEmployees(employeeData, 3500))
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model = KNeighborsClassifier(n_neighbors=4)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = cross_val_score(model,x_test,y_test,cv=5)
```
