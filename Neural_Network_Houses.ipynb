```python
import pandas as pd

house_df = pd.read_csv('/content/drive/MyDrive/CL/Day_3/Kopia kc_house_data_preprocessed.csv')
```

```python
house_df
house_df.dropna()
```

```python
house_df.info()

```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
from sklearn.model_selection import train_test_split
X = house_df.drop(['price'],axis=1)
y = house_df['price']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
X_train  = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
```

```python
nn_model = tf.keras.models.Sequential()
```

```python
from tensorflow.keras.layers import Dense
nn_model.add(Dense(X_train.shape[1], input_shape=(X_train.shape[1],), activation='relu'))

# Dodanie kolejnych warstw ukrytych
nn_model.add(Dense(64, activation='relu'))
nn_model.add(Dense(32, activation='relu'))
# Wyjsciowa
nn_model.add(Dense(1,activation='relu'))
```

```python
nn_model.compile(
    optimizer='Adam',loss='mean_squared_error'

)
```

```python
nn_model.fit(X_train,y_train,batch_size=32,epochs=100,validation_split=0.15)
```

```python
y_pred = nn_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
