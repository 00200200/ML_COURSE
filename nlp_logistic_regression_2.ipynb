```python
df = pd.read_csv('/content/drive/MyDrive/CL/Day 4/imdb_reviews_preprocessed_2.fixed.csv')
df
```

```python
df
df.dropna(subset=['review_lemmatized'], inplace=True)
df['label'].describe()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=10000)
```

```python
X_tfidf = tfidf_vectorizer.fit_transform(df['review_lemmatized'])
X_tfidf_vector = X_tfidf.toarray()
df_tfidf = pd.DataFrame(X_tfidf_vector, columns=tfidf_vectorizer.get_feature_names_out())
```

```python
df_tfidf['label'] = df['label'].apply(lambda sentiment: 1 if sentiment == 'pos' else 0)
y = df['label']
```

```python
df_tfidf
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_tfidf,y,test_size=0.2)

```

```python
from sklearn.linear_model import LogisticRegression

logreg_model = LogisticRegression(max_iter=1000)
logreg_model.fit(X_train, y_train)
```

```python
y_pred_logreg = logreg_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred_logreg))
```
