```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download("wordnet")
lemmatizer = WordNetLemmatizer()
def format_text(line):
  words = line.split()
  lemmatized_words = [lemmatizer.lemmatize(token) for token in words]
  line = " ".join(lemmatized_words)
  return line
```

```python
df["review_lemmatized"] = df["review_cleaned"].apply(format_text)
```
