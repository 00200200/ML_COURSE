```python
from google.colab import drive
drive.mount('/drive')
path_to_file = '/drive/MyDrive/second_task.csv'
import pandas as pd
df = pd.read_csv(path_to_file)
df
```

```python
from sklearn.preprocessing import StandardScaler
standarized_method = StandardScaler()
columns_ts = ['age', 'restbp', 'chol', 'restecg', 'maxhr', 'oldpeak', 'slope', 'ca']
heart_data_standardized = standarized_method.fit_transform(df[columns_ts])
heart_data_standardized
heart_data_standardized = pd.DataFrame(heart_data_standardized,columns=columns_ts)
heart_data_standardized = pd.concat([heart_data_standardized,df.drop(columns=columns_ts)],axis=1)
heart_data_standardized.to_csv('/drive/MyDrive/standarized_hear_data.csv',index=False)
```

```python
from sklearn.preprocessing import MinMaxScaler
normalizer = MinMaxScaler()
heart_data_normalized = normalizer.fit_transform(df[columns_ts])
heart_data_normalized
```

```python
heart_data_normalized = pd.DataFrame(heart_data_normalized,columns=columns_ts)
heart_data_normalized = pd.concat([heart_data_normalized,df.drop(columns=columns_ts)],axis=1)
heart_data_normalized.to_csv('/drive/MyDrive/normalized_hear_data.csv',index=False)
heart_data_normalized
```
