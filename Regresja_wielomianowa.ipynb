```python
path_to_file = '/drive/MyDrive/diamonds_preprocessed.csv'
```

```python
import pandas as pd

```

```python
diamonds_df = pd.read_csv(path_to_file)
```

```python
diamonds_df
```

```python
from sklearn.model_selection import train_test_split
X = diamonds_df.drop(['price'],axis=1)
y = diamonds_df['price']
```

```python
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
scaler = StandardScaler()
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np
def print_metrics(y_true, y_pred, ):
    r2 = r2_score(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)

    print(f"R2: {r2}")
    print(f"MAE: {mae}")
    print(f"MSE: {mse}")
    print(f"RMSE: {rmse}")
    print("-" * 30)
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
x_train_poly_scaled = scaler.fit_transform(x_train_poly)
x_test_poly_scaled = scaler.transform(x_test_poly)
```

```python
model_poly = LinearRegression()
model_poly.fit(x_train_poly_scaled,y_train)
```

```python
y_pred_train_poly = model_poly.predict(x_train_poly_scaled)
y_pred_test_poly = model_poly.predict(x_test_poly_scaled)
```

```python
print_metrics(y_train,y_pred_train_poly)
```

```python
print_metrics(y_test,y_pred_test_poly)
```

```python
poly = PolynomialFeatures(degree=3)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
x_train_poly_scaled = scaler.fit_transform(x_train_poly)
x_test_poly_scaled = scaler.transform(x_test_poly)
model_poly = LinearRegression()
model_poly.fit(x_train_poly_scaled,y_train)
y_pred_train_poly = model_poly.predict(x_train_poly_scaled)
y_pred_test_poly = model_poly.predict(x_test_poly_scaled)
```

```python
print_metrics(y_train,y_pred_train_poly)
```

```python
print_metrics(y_test,y_pred_test_poly)
```
